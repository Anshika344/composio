Types of object in python


1.   Immutable object - we cannot change the values once assigned.
2.   Mutable object - we can change the values once created.

Immutable objects
1. int
2. float
3. string
4. bool
5. tuple

Mutable objects
1. list
2. set
3. dictionary




#**List**

```python
#list should be included in the square brackets
my_list = [2,4,5,8]
print(my_list)
type(my_list)
```

```python
#list can contain multiple data types
my_list = [2,3,6.4,'anshika', True]
print(my_list)
```

```python
#lists are mutable --> changeable
#add element to the list using append
my_list = [1, 4, 6.6, 'anshika', True]
my_list.append(10)
print(my_list)
```

```python
#print element using their index
my_list = [1, 4, 6.6, 'anshika', True]
print(my_list[0])
print(my_list[2])
print(my_list[4])
```

```python
#list allows duplicate element
list_1 = [1, 2, 3, 1, 2, 5, 6, 3, 67, 8]
print(list_1)
```

```python
#for counting length of list or total nos of elements in list
print(len(list_1))
```

```python
#initiating an empty list
list_2 = []
print(list_2)
```

```python
#join the two lists (concatenation)
list_1 = [4, 5, 6]
list_2 = [1, 5, 7]
list_3 = list_1 + list_2
print(list_3)
```

```python
#deleting an element
print(list_3)
del list_3[3]
print(list_3)
```

# **Tuples**

```python
tuple_1= (7, 4 , 3 , 2)
print(tuple_1)
type(tuple_1)
```

```python
#tuple allows multiple datasets
tuple_2 = (2, 4.5 , 56 , 'ash', False)
print(tuple_2)
```

```python
#converting a list into tuple
my_list =[2,3,4.5,'ash', 8]
print(my_list)

my_tuple = tuple(my_list)
print(my_tuple) # we can see the difference between brackets
```

```python
# tuples are immutable --> no changeable
my_tuple = (1,2,3,4,5)
my_tuple.append[7]
print(my_tuple)
```

```python
print(len(my_tuple))
```

# **Set**
curly braces

```python
my_set = {2,4,5,6,7}
print(my_set)
type(my_set)
```

```python
#set doesnot involve indexing
my_set ={ 2, 4, 5, 6, 7}
print(my_set[0])
```

```python
#converting list into set
my_list = [2, 4, 5, 6]
print(my_list)
x = set(my_list)
print(x)
```

```python
# set doesnot allow duplicate
my_set = {2, 4, 5, 6, 2, 4}
print(my_set)
```

#**Dictionary**

 key value pairs

```python
my_dictionary ={'name':'ash', 'age': 20 , 'city':'patna', 'roll no':15}
print(my_dictionary)
type(my_dictionary)
```

```python
print(my_dictionary['name'])
print(my_dictionary['age'])
print(my_dictionary['city'])
print(my_dictionary['roll no'])
```

```python
#dictionaries does not allow duplicate
my_dictionary ={'name':'ash', 'age': 20 , 'city':'patna', 'roll no':15 , 'name':'ash'}
print(my_dictionary)
```
